Escrever um programa que faz uma contagem regressiva de 10 até 0. Imprimir uma mensagem de "acabou!" no final.

```python

palavra = input("Digite um número: ")
numero = int(palavra)
for i in range(numero, -1, -1):
  print(i)
  if i == 0:
    print("Acabou!")
```

Escrever um programa que recebe as notas dos 04 brimestres e imprime a média com o resultado de aprovado, recuperção ou reprovado

Regras:
utilizar while;
Nota igual ou maior que 7 aprovados
Nota igual ou maior que 5 recuperação;
Nota menor que 5 reprovado

```python
soma = 0
contador = 1

while contador <= 4:
    nota = float(input(f"Digite a nota do {contador}º bimestre: "))
    soma += nota
    contador += 1

media = soma / 4

print(f"Média final: {media}")

if media >= 7:
    print("Você está: Aprovado")
elif media >= 5:
    print("Você está: Recuperação")
else:
    print("Você está: Reprovado")
```

Projeto caixa eletrônico
Você foi contratado para desenvolver um sistema simples de caixa eletrônico
O sistema deverá funcionar através de um menu interativo utilizando "while"
O programa devera iniciar com um saldo inicial de R$1000,00 e permitir que o usuário escolha difirentes operações até decidir sair do sistema. Sacar dinheiro, ver saldo, depositar dinheiro ou sair do banco

```python
saldo = 1000.0

while True:
    print(f"Saldo: R$ {saldo}")
    print("1-Sacar | 2-Depositar | 3-Sair")
    opcao = input("Opção: ")

    if opcao == "1":
        saldo -= float(input("Valor: "))

    elif opcao == "2":
        saldo += float(input("Valor: "))

    elif opcao == "3":
        break

```

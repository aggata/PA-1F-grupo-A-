Calculadora com as 4 operações
Menu: somar, subtrair, multitiplicar, divisão, sair

```python
print("___________Calculadora___________")
print("opção 1: somar")
print("opção 2: subtrair")
print("opção 3: multiplicar")
print("opção 4: dividir")
print("opção 5: sair")
print("_________________________________")

opcao = int(input("Digite a opção desejada: "))

if opcao == 1:
    num1 = float(input("Digite o primeiro número: "))
    num2 = float(input("Digite o segundo número: "))
    resultado = num1 + num2
    print(f"O resultado é: {resultado}")

elif opcao == 2:
    num1 = float(input("Digite o primeiro número: "))
    num2 = float(input("Digite o segundo número: "))
    resultado = num1 - num2
    print(f"O resultado é: {resultado}")

elif opcao == 3:
    num1 = float(input("Digite o primeiro número: "))
    num2 = float(input("Digite o segundo número: "))
    resultado = num1 * num2
    print(f"O resultado é: {resultado}")

elif opcao == 4:
    num1 = float(input("Digite o primeiro número: "))
    num2 = float(input("Digite o segundo número: "))
    if num2 != 0:
        resultado = num1 / num2

    else:
        resultado = "Erro: Não é possível dividir por zero!"
    print(f"O resultado é: {resultado}")

elif opcao == 5:
    print("Saindo da calculadora...")
```

Jogo inteligente
1- par/impar
2- positivo
3- maior de 02 números
4- sair

```python
print("___________Jogo Inteligente___________")
print("opção 1: par/impar")
print("opção 2: positivo")
print("opção 3: maior de 02 números")
print("opção 4: sair")
print("_______________________________________")

opcao = int(input("Digite a opção desejada: "))

if opcao == 1:
    num = int(input("Digite um número: "))
    if num % 2 == 0:
        print("O número é par!")
    else:
        print("O número é ímpar!")

elif opcao == 2:
    num = int(input("Digite um número: "))
    if num > 0:
        print("O número é positivo!")
    elif num < 0:
        print("O número é negativo!")
    else:
        print("O número é zero!")

elif opcao == 3:
    num1 = int(input("Digite o primeiro número: "))
    num2 = int(input("Digite o segundo número: "))
    if num1 > num2:
        print("O primeiro número é maior!")
    elif num1 < num2:
        print("O segundo número é maior!")
    else:
        print("Os números são iguais!")

elif opcao == 4:
    print("Saindo do jogo...")
```

Escrever um programa que verifica a senha, o usuário terá três tentativas, caso erre as três as três imprimir a mensagem "acesso negado e tente novamente mais tarde!" e encerrar o programa

```python
senha = ""
tentativas = 3


while senha != "1111" and tentativas > 0:
    senha = input("Digite a senha: ")
    tentativas -= 1

    if senha == "1111":
        print("Acesso permitido!")

    elif tentativas == 0:
        print("Acesso negado! Tente novamente mais tarde!")
    else:
        print(f"Acesso negado! Você ainda tem {tentativas} tentativas.")
```

Escrever um programa que recebe o nome e a idade do usuário e verifica se o usuário é criança, adolescente ou adulto

>= 12 anos é criança
>= 17 adolescente
>= 18 adulto

```python
# Entrada

nome = str(input("digite seu nome: "))
idade = int(input("digite sua idade: "))

# Processamento / Saida

if idade <= 12:
  print(f"{nome}, você tem {idade} anos de idade, você é criança.")

elif idade <= 17:
  print(f"{nome}, você tem {idade} anos de idade, você é adolescente.")

else:
  print(f"{nome}, você tem {idade} anos de idade, você é adulto.")
```

Em um jogo, um personagem possui uma variavel chamada energia. O programa deve: Se a energia for maior que 70 - Forte, se a energia estiver entre 30 e 70 - Cansado, se a energia estiver menorr que 30 - Fraco

```python
# Entrada

Energia = int(input("digite a energia do seu personagem: "))

# Processamento

if Energia <= 30:
  print("Ele está fraco.")

elif Energia <= 70:
  print("Ele está cansado.")

else:
  print("Ele está cansado")
```

Escreva um programa que classifica os tipos de triângulo e se é um trângulo. Validar se os lados digitados pode formar um trângulo

```python
# Entrada

lado1 = int(input("Digite o primeiro lado: "))
lado2 = int(input("Digite o segundo lado: "))
lado3 = int(input("Digite o terceiro lado: "))

# Processamento/ Saida

if lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1:
  print("É um triângulo")

if lado1 == lado2 == lado3:
  print("Triângulo equilátero")

elif lado1 != lado2 and lado2 != lado3:
  print("Triângulo escaleno.")

elif lado1 == lado2 and lado1 == lado3:
  print("Triângulo isósceles.")

else:
    print("Não é um triângulo inválido.")
```

Escrever um programa que calcula a media das notas do aluno

Entrada: nome, idade, as 4 notas do bimestre

Processamento: calcular a media

Saida: mostrar o nome do aluno e a média obtida

```python
# Entrada

nome = str (input("Digite o nome do aluno: "))
idade = int (input("Digite a idade do aluno:"))
nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota:"))
nota3 = float(input("Digite a terceira nota: "))
nota4 = float(input("Digite a quarta nota: "))

# Processamento

media = (nota1 + nota2 + nota3 + nota4)/ 4

# Saida

print (f"a media da aluna {nome}, com {idade} anos é: {media}")
```

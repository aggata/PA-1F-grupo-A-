Solicite ao usuário um número inteiro positivo.

Utilizando um laço de repetição, calcule e exiba a soma de
todos os números de 1 até o número informado.

Exemplo:

Digite um número: 5

Resultado: 15

Pois:

1 + 2 + 3 + 4 + 5 = 15

```python
numero = int(input("Digite um número inteiro positivo: "))
soma = 0
for i in range(1, numero + 1):
    soma += i
print(f"Resultado: {soma}")
```

seu fatorial utilizando um laço de repetição.

Solicite ao usuário o valor do saque e informe a menor
quantidade possível de notas para realizar o pagamento.

Exemplo:

Digite o valor do saque: 287

Saída esperada:

2 nota(s) de R$100
1 nota(s) de R$50
1 nota(s) de R$20
1 nota(s) de R$10
1 nota(s) de R$5
1 nota(s) de R$2


```python
valor = int(input("Digite o valor do saque: "))

nota = 100

while nota >= 1:
    quantidade = valor // nota

    if quantidade > 0:
        print(quantidade, "nota(s) de R$", nota)

    valor = valor % nota

    if nota == 100:
        nota = 50
    elif nota == 50:
        nota = 20
    elif nota == 20:
        nota = 10
    elif nota == 10:
        nota = 5
    elif nota == 5:
        nota = 2
    elif nota == 2:
        nota = 1
    else:
        break
```

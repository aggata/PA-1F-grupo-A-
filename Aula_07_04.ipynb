Crie um programa que leia a idade de uma pessoa e classifique conforme as regras abaixo:
	•	Até 17 anos: Menor de idade
	•	De 18 a 59 anos: Adulto
	•	60 anos ou mais: Idoso


```python
# Entrada

nome = str(input("digite seu nome: "))
idade = int(input("digite sua idade: "))

# Processamento / Saida

if idade <= 17:
  print(f"{nome}, você tem {idade} anos de idade, você é menor de idade.")

elif idade <= 59:
  print(f"{nome}, você tem {idade} anos de idade, você é adulto.")

else:
  print(f"{nome}, você tem {idade} anos de idade, você é adulto.")
```

Crie um programa que leia o valor de uma compra e aplique o desconto conforme as regras:
	•	Até R$100: sem desconto
	•	De R$101 a R$500: 10% de desconto
	•	Acima de R$500: 20% de desconto
Ao final, mostre o valor final da compra.


```python
# Entrada

valor = float(input("digite o valor da compra: "))

# Processamento / Saida

if valor <= 100:
  print(f"O valor final da compra é de R${valor}.")

elif valor <= 500:
  print(f"O valor final da compra é de R${valor * 0.9}.")

else:
  print(f"O valor final da compra é de R${valor * 0.8}.")
```

Crie um programa que leia três valores e verifique se eles podem formar um triângulo.
Para isso, a soma de dois lados deve ser maior que o terceiro.
Mostre uma mensagem informando se é ou não um triângulo válido.


```python
# Entrada

lado1 = int(input("Digite o primeiro lado: "))
lado2 = int(input("Digite o segundo lado: "))
lado3 = int(input("Digite o terceiro lado: "))

# Processamento/ Saida

if lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1:
  print("É um triângulo válido.")

if lado1 == lado2 == lado3:
  print("Triângulo equilátero")

elif lado1 != lado2 and lado1 != lado3:
  print("Triângulo escaleno.")

elif lado1 == lado2 and lado1 == lado3:
  print("Triângulo isósceles.")

else:
  print("Triângulo inválido.")
```

Crie um programa que leia a energia de um personagem (de 0 a 100) e classifique:
	•	0: Morto
	•	1 a 30: Crítico

	•	71 a 100: Saudável

Crie um programa que leia a média final de um aluno e classifique:
	•	Média maior ou igual a 7: Aprovado
	•	Média entre 5 e 6.9: Recuperação


```python
# Entrada

nome = str (input("Digite o nome do aluno: "))
idade = int (input("Digite a idade do aluno:"))
nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))
nota3 = float(input("Digite a terceira nota: "))
nota4 = float(input("Digite a quarta nota: "))

# Processamento/ Saida

media = (nota1 + nota2 + nota3 + nota4) / 4

if media >= 7:
  print(f"O aluno {nome} está aprovado com média {media}.")

elif media >= 5:
  print(f"O aluno {nome} está em recuperação com média {media}.")

else:
  print(f"O aluno {nome} está reprovado com média {media}.")


```

Crie um programa que leia o salário de um funcionário e sua pontuação de desempenho. O programa deve calcular um bônus de acordo com as regras:
	•	Pontuação menor que 5: sem bônus
	•	Pontuação de 5 a 7: bônus de 5% do salário
	•	Pontuação de 8 a 9: bônus de 10% do salário
	•	Pontuação igual a 10: bônus de 15% do salário

Ao final, mostre o valor do bônus e o salário final com o bônus aplicado.Média menor que 5: Reprovado

```python
# Entrada
pontuação = int(input("Digite a pontuação do funcionário: "))
salario = float(input("Digite o salário do funcionário: "))

# Pontuação

if pontuação <= 5:
  print("O funcionário está reprovado.")

if pontuação <= 7:
  bonus = 5 / 100 * salario
  salario * 0.05
  print(f"O funcionário receberá um bônus de R${bonus}.")

if pontuação <= 9:
  bonus = 10 / 100 * salario
  salario * 0.1
  print(f"O funcionário receberá um bônus de R${bonus}.")


if pontuação == 10:

  bonus = 15 / 100 * salario
  salario * 0.15
  print(f"O funcionário receberá um bônus de R${bonus}")

elif pontuação >= 10:
  print("Pontuação inválida.")

elif pontuação <= 5:
  print("O funcionário não receberá bônus.")

else:
  print("O funcionário não receberá bônus.")
```

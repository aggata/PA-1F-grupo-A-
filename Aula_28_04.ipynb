 Escrever um programa que o usuario digita uma palavra e o programa verifica se a palavra é palindroma e imprime a palavra

```python
palavra = input("Digite uma palavra: ")
invertida = ""

for i in (palavra):

  invertida = i + invertida

if palavra == invertida:
  print("A palavra é palindroma")

elif palavra != invertida:
  print("A palavra não é palindroma")

else:
  print("A palavra não é palindroma")
```
